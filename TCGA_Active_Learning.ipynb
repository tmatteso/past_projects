```python


from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest # needs a y

seen_set = set()
all_files = dict()
data = np.empty([1,60485])


target_dir = "/content/drive/MyDrive/RNAseq_files/"
counter = 0
for root, dirs, files in os.walk(target_dir):
    for file in files: 
        if file.endswith(".txt"):
          #print(file)
          if file in f_name_2_case_id.keys() and f_name_2_case_id[file] in simple_map:
            #print("poop")
            counter += 1
            file_in_dir = target_dir + file
            gene_names = []
            expression_values = []
            with open(file_in_dir) as f:
                #print("shloop")
                line = f.readline().strip().split("\t")
                expression_values.append(float(line[1]))
                while line:
                    line = f.readline().strip().split("\t")
                    if len(line) == 2:
                        expression_values.append(float(line[1]))
                    else:
                        break
                # apply the normalization here across all the expression values
                expression_values = [1.0 if i < 1 else i for i in expression_values]
                expression_values = np.log2(np.array(expression_values))
                # add label 
                # now we change things a bit 
                row = new_df.iloc[np.where(new_df["filename"] == file)[0][0]]
                # print(np.where(new_df["filename"] == file)[0][0])
                #row = row[["vital_status", "days_to_birth", "days_to_death", "days_to_treatment_start", "days_to_treatment_end"]]
                row = row[[ "days_to_death", "days_to_treatment_end"]]
                #print(row)
                row = (row.to_numpy())
                new_expression_values = np.append(np.array(expression_values),row)
                new_expression_values = new_expression_values[:,np.newaxis].T
                # print(file)
                # print(data.shape, "data")
                # print(new_expression_values.shape, "expression")
                if new_expression_values.shape[1] != 60485: print(new_expression_values[0, -10:])
                # turn all floats less than 1 to 1 and log2 transform
                #expression_values = np.log2(np.array(expression_values))
                data = np.vstack((data,new_expression_values))
            if (counter) % 100 == 0:
              print(counter)
        #raise KeyError

print(data.shape)
```
